Using Percentiles

```python
lower, upper = df.price.quantile([0.001,0.999])

#Removing the top and bottom 0.1% observations
df[(df['price_per_sqft']>lower) & (df['price_per_sqft']<upper)]
```

```python
#Plotting the distribution after removal of outliers
sns.displot(df['price_per_sqft'],kde = True)
```

```python
mean = df.price_per_sqft.mean()
mean
```

```python
std_deviation = df.price_per_sqft.std()
std_deviation
```

#Using Standard Deviation 

```python
#Removing outliers greater than 4 Standard Deviations from the mean
df_new=df[(df['price_per_sqft']>(mean-4*std_deviation)) & (df['price_per_sqft']<(mean+4*std_deviation))]
```

```python
sns.displot(df_new['price_per_sqft'],kde = True)
```

#Using Z Score

```python
#Calculating Z Score
df['zscore']= (df['price_per_sqft']-mean/std_deviation)
```

```python
#Removing Outliers by exludsing observations having z score greater than 4
df_new1= df[(df['zscore']>-4) & (df['zscore']<4)]
df_new1
```
