# Outlier Detection and Removal using Percentile

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("Desktop\height.csv")
df.head()
```

Detect Outliers using percentile

```python
max_threshold = df['Height'].quantile(0.95)
max_threshold
```

```python
df[df['Height']>max_threshold]
```

```python
min_threshold = df['Height'].quantile(0.05)
min_threshold
```

```python
df[df['Height']<min_threshold]
```

#Remove Outliers

```python
df[(df['Height']>min_threshold) & (df['Height']<max_threshold)]
```

#Bangalore House Prices - Outlier Detection and Removal using Percentile

```python
df_b = pd.read_csv("Desktop/bhp.csv") 
df_b
```

```python
df_b.shape
```

```python
df_b.describe()
```

Explore samples that are above 99.90% percentile and below 1% percentile rank

```python
min_threshold, max_threshold = df_b.price_per_sqft.quantile([0.001,0.999])
min_threshold, max_threshold
```

Remove outliers

```python
df2 = df_b[(df_b['price_per_sqft']>min_threshold) & (df_b['price_per_sqft']<max_threshold)]
df2.shape
```

```python
df2.describe()
```
